```python
import pandas as pd
import numpy as np
import psycopg2
import sys, os
import pandas.io.sql as psql
import sqlalchemy  # Package for accessing SQL databases via Python
```

```python
f = open('db.txt', 'r')
host = f.readline().strip()
dbname = f.readline().strip()
user = f.readline().strip()
password = f.readline().strip()
schema = f.readline().strip()
database = f.readline().strip()
f.close()
```

```python
# Connect to database (Note: The package psychopg2 is required for Postgres to work with SQLAlchemy)

engine = sqlalchemy.create_engine("postgresql://{}:{}@{}/{}".format(user, password, host, database))
con = engine.connect()

# Verify that there are no existing tables
print(engine.table_names())
```

```python
table = "measures"
schema = "public" #always public

# Command
sql_command = "SELECT * FROM {}.{};".format(str(schema), str(table))
print (sql_command)
measures_df = pd.read_sql(sql_command,
                     con=engine,
                    index_col="index")

print(measures_df.shape)
```
